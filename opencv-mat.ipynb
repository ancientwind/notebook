## Accessing and Modifying pixel values and Properties

```python
import cv2 as cv

img = cv.imread('img/batman.jpg')
print(img.shape)
px = img[100,200] # Blue, Green, Red value at pixel (100,200)
print(px)

blue = img[100, 200, 0]
print(blue)

# better pixel accessing and editing method:

px2 = img.item(100, 200, 0)
print(px2)

img.itemset((100,200,0), 100) # set (100,200,0) to value 100
print(img.item(100, 200, 0))

# Image properties include number of rows, columns and channels, type of image data, number of pixels etc.
print(type(img), img.size)
print(img.dtype)
```

## Making Borders for Images (Padding)

```python
import cv2 as cv
import matplotlib.pyplot as plt

BLUE = [255, 0, 0]

img = cv.imread('img/opencv.jpeg')
print('img: ', img.shape)

replicate = cv.copyMakeBorder(img, 10, 10, 10, 10, cv.BORDER_REPLICATE)
print('replicate: ', replicate.shape)

reflect = cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_REFLECT)
reflect101 = cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_REFLECT_101)
wrap = cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_WRAP)
constant= cv.copyMakeBorder(img,10,10,10,10,cv.BORDER_CONSTANT,value=BLUE)

plt.subplot(231), plt.title('ORIGINAL'), plt.imshow(img, 'gray')
plt.subplot(232), plt.title('REPLICATE'), plt.imshow(replicate, 'gray')
plt.subplot(233),plt.imshow(reflect,'gray'),plt.title('REFLECT')
plt.subplot(234),plt.imshow(reflect101,'gray'),plt.title('REFLECT_101')
plt.subplot(235),plt.imshow(wrap,'gray'),plt.title('WRAP')
plt.subplot(236),plt.imshow(constant,'gray'),plt.title('CONSTANT')
```
